```python
#in this model we will predict movie genres using movie plots
#to predict we are using bernoulli naive bayes algorithm
#in this section we import all the required libraries
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

#we import movie database
n1 = LabelEncoder()
df = pd.read_csv('movies_metadata.csv')
```

```python
#here is the list of all the movie genres considered in this model
genre_types_temp = ['Drama', 'Comedy', 'Adventure', 'History', 'War', 'Thriller', 'Crime', 'Fantasy', 'Horror', 'Family', 'Documentary', 'Mystery', 'Romance', 'Science Fiction', 'Action']
```

```python
#now we will count the number of movies in each genre
genre_count = {'Drama':0, 'Comedy':0, 'Adventure':0, 'History':0, 'War':0, 'Thriller':0, 'Crime':0, 'Fantasy':0, 'Horror':0, 'Family':0, 'Documentary':0, 'Mystery':0, 'Romance':0, 'Science Fiction':0, 'Action':0}
for i in df['genres']:
    t = i.split(', {')
    for j in range(len(t)):
        q = t[j].split("'name':")
        if(len(q)<=1):
            continue
        q = q[1]
        q = q.split("'")
        q = q[1]
        if q in genre_types_temp:
            genre_count[q] = genre_count[q] + 1
print(genre_count)
```

```python
#in this section we vectorize the movie subplots from the database
data = []
for i in range(len(df)):
    data.append(df['overview'][i])
tfidf = TfidfVectorizer(lowercase=True, analyzer='word', stop_words='english', ngram_range=(1,1), use_idf=True)
data = np.array(data)
data = tfidf.fit_transform(data)
```

```python
#this section contains input for our model
t = "After Thanos, an intergalactic warlord, disintegrates half of the universe, the Avengers must reunite and assemble again to reinvigorate their trounced allies and restore balance."
t = np.array(t).reshape(-1,1)
t = tfidf.fit_transform(t[0])
```

```python
#in this section we train our model
#we train each genre model separately

genre_ans = []
#Drama_genre
Drama = BernoulliNB()
Drama_data = np.array(Drama_data)
Drama_out = np.array(Drama_out)
X_Drama = tfidf.transform(Drama_data)
Y_Drama = Drama_out
Drama.fit(X_Drama, Y_Drama)
if (Drama.predict(t)[0]==1):
    genre_ans.append("Drama")

#Comedy_genre
Comedy = BernoulliNB()
Comedy_data = np.array(Comedy_data)
Comedy_out = np.array(Comedy_out)
X_Comedy = tfidf.transform(Comedy_data)
Y_Comedy = Comedy_out
Comedy.fit(X_Comedy, Y_Comedy)
if (Comedy.predict(t)[0]==1):
    genre_ans.append("Comedy")

#Adventure_genre
Adventure = BernoulliNB()
Adventure_data = np.array(Adventure_data)
Adventure_out = np.array(Adventure_out)
X_Adventure = tfidf.transform(Adventure_data)
Y_Adventure = Adventure_out
Adventure.fit(X_Adventure, Y_Adventure)
if (Adventure.predict(t)[0]==1):
    genre_ans.append("Adventure")

#History_genre
History = BernoulliNB()
History_data = np.array(History_data)
History_out = np.array(History_out)
X_History = tfidf.transform(History_data)
Y_History = History_out
History.fit(X_History, Y_History)
if (History.predict(t)[0]==1):
    genre_ans.append("History")

#War_genre
War = BernoulliNB()
War_data = np.array(War_data)
War_out = np.array(War_out)
X_War = tfidf.transform(War_data)
Y_War = War_out
War.fit(X_War, Y_War)
if (War.predict(t)[0]==1):
    genre_ans.append("War")

#Thriller_genre
Thriller = BernoulliNB()
Thriller_data = np.array(Thriller_data)
Thriller_out = np.array(Thriller_out)
X_Thriller = tfidf.transform(Thriller_data)
Y_Thriller = Thriller_out
Thriller.fit(X_Thriller, Y_Thriller)
if (Thriller.predict(t)[0]==1):
    genre_ans.append("Thriller")

#Crime_genre
Crime = BernoulliNB()
Crime_data = np.array(Crime_data)
Crime_out = np.array(Crime_out)
X_Crime = tfidf.transform(Crime_data)
Y_Crime = Crime_out
Crime.fit(X_Crime, Y_Crime)
if (Crime.predict(t)[0]==1):
    genre_ans.append("Crime")

#Fantasy_genre
Fantasy = BernoulliNB()
Fantasy_data = np.array(Fantasy_data)
Fantasy_out = np.array(Fantasy_out)
X_Fantasy = tfidf.transform(Fantasy_data)
Y_Fantasy = Fantasy_out
Fantasy.fit(X_Fantasy, Y_Fantasy)
if (Fantasy.predict(t)[0]==1):
    genre_ans.append("Fantasy")

#Horror_genre
Horror = BernoulliNB()
Horror_data = np.array(Horror_data)
Horror_out = np.array(Horror_out)
X_Horror = tfidf.transform(Horror_data)
Y_Horror = Horror_out
Horror.fit(X_Horror, Y_Horror)
if (Horror.predict(t)[0]==1):
    genre_ans.append("Horror")

#Family_genre
Family = BernoulliNB()
Family_data = np.array(Family_data)
Family_out = np.array(Family_out)
X_Family = tfidf.transform(Family_data)
Y_Family = Family_out
Family.fit(X_Family, Y_Family)
if (Family.predict(t)[0]==1):
    genre_ans.append("Family")

#Documentary_genre
Documentary = BernoulliNB()
Documentary_data = np.array(Documentary_data)
Documentary_out = np.array(Documentary_out)
X_Documentary = tfidf.transform(Documentary_data)
Y_Documentary = Documentary_out
Documentary.fit(X_Documentary, Y_Documentary)
if (Documentary.predict(t)[0]==1):
    genre_ans.append("Documentary")

#Mystery_genre
Mystery = BernoulliNB()
Mystery_data = np.array(Mystery_data)
Mystery_out = np.array(Mystery_out)
X_Mystery = tfidf.transform(Mystery_data)
Y_Mystery = Mystery_out
Mystery.fit(X_Mystery, Y_Mystery)
if (Mystery.predict(t)[0]==1):
    genre_ans.append("Mystery")

#Romance_genre
Romance = BernoulliNB()
Romance_data = np.array(Romance_data)
Romance_out = np.array(Romance_out)
X_Romance = tfidf.transform(Romance_data)
Y_Romance = Romance_out
Romance.fit(X_Romance, Y_Romance)
if (Romance.predict(t)[0]==1):
    genre_ans.append("Romance")

#ScienceFiction_genre
ScienceFiction = BernoulliNB()
ScienceFiction_data = np.array(ScienceFiction_data)
ScienceFiction_out = np.array(ScienceFiction_out)
X_ScienceFiction = tfidf.transform(ScienceFiction_data)
Y_ScienceFiction = ScienceFiction_out
ScienceFiction.fit(X_ScienceFiction, Y_ScienceFiction)
if (ScienceFiction.predict(t)[0]==1):
    genre_ans.append("Science Fiction")

#Action_genre
Action = BernoulliNB()
Action_data = np.array(Action_data)
Action_out = np.array(Action_out)
X_Action = tfidf.transform(Action_data)
Y_Action = Action_out
Action.fit(X_Action, Y_Action)
if (Action.predict(t)[0]==1):
    genre_ans.append("Action")
```
